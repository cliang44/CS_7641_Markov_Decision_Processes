```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Easy Grid
```

```python
EG_FILE = ['Easy Value.csv',
           'Easy Policy.csv',
           'Easy Q-Learning L0.1 E0.2.csv',
           'Easy Q-Learning L0.5 E0.2.csv',
           'Easy Q-Learning L0.9 E0.2.csv',
           'Easy Q-Learning L0.1 E0.8.csv',
           'Easy Q-Learning L0.5 E0.8.csv',
           'Easy Q-Learning L0.9 E0.8.csv'
            ]
HG_FILE = ['Hard Value.csv',
           'Hard Policy.csv',
           'Hard Q-Learning L0.1 E0.2.csv',
           'Hard Q-Learning L0.5 E0.2.csv',
           'Hard Q-Learning L0.9 E0.2.csv',
           'Hard Q-Learning L0.1 E0.8.csv',
           'Hard Q-Learning L0.5 E0.8.csv',
           'Hard Q-Learning L0.9 E0.8.csv'
            ]
```

```python
legend1 = ['Value Iteration', 
          'Policy Iteration',
          'Q-Learning Iteration']
legend2 = ['L0.1 E0.2',
           'L0.5 E0.2',
           'L0.9 E0.2',
           'L0.1 E0.8',
           'L0.5 E0.8',
           'L0.9 E0.8']
```

```python
FILES=[EG_FILE[2],EG_FILE[3],EG_FILE[4],EG_FILE[5],EG_FILE[6],EG_FILE[7]]
df = pd.DataFrame()
i = 0
for file_ in FILES:
    file_df = pd.read_csv(file_,index_col=False, header=0)
    file_df['learning rate/epsilon'] = str(legend2[i])
    df = df.append(file_df)
    i = i + 1
df.head() 
df = df.drop(columns=['policy'])
df.head()
```

```python
df2 = df[df['iter'] > 500]
df2.head()
plt.figure(1)
sns.catplot(x="learning rate/epsilon", y="reward",  kind="bar", data=df2,palette = "pastel")
plt.figure(2)
sns.catplot(x="learning rate/epsilon", y="steps",  kind="bar", data=df2,palette = "pastel")
plt.figure(3)
sns.catplot(x="learning rate/epsilon", y="convergence",  kind="bar", data=df2,palette = "pastel")
plt.figure(4)
sns.catplot(x="learning rate/epsilon", y="time",  kind="bar", data=df2,palette = "pastel")
```

```python
FILES=[EG_FILE[0],EG_FILE[1],EG_FILE[5]]
df = pd.DataFrame()
i = 0
for file_ in FILES:
    file_df = pd.read_csv(file_,index_col=False, header=0)
    file_df['Algorithm'] = str(legend1[i])
    df = df.append(file_df)
    i = i + 1
df.head() 
df2 = df.copy()
df = df[df['iter'] <= 100]
df = df.drop(columns=['policy'])
df.head()
```

```python
plt.figure(1)
sns.relplot(x="iter", y="reward", hue = "Algorithm", kind="line", data=df,legend = False,palette = "pastel")
plt.title('Easy Grid World Reward')
plt.ylim(0, 100)
plt.legend(legend1)

plt.figure(2)
sns.relplot(x="iter", y="steps", hue = "Algorithm", kind="line", data=df,legend = False,palette = "pastel")
plt.title('Easy Grid World Steps')
plt.ylim(0, 100)
plt.legend(legend1)

plt.figure(3)
sns.relplot(x="iter", y="time", hue = "Algorithm", kind="line", data=df,legend = False,palette = "pastel")
plt.title('Easy Grid World Time')
#plt.ylim(0, 100)
plt.legend(legend1)

plt.figure(4)
sns.relplot(x="iter", y="convergence", hue = "Algorithm", kind="line", data=df,legend = False,palette = "pastel")
plt.title('Easy Grid World Convergence')
plt.legend(legend1)


plt.figure(5)
sns.relplot(x="iter", y="convergence", hue = "Algorithm", kind="line", 
            data=df2[df2['Algorithm'] == "Q-Learning Iteration"],legend = False,palette = "Set2")
plt.title('Easy Grid World Q-learning Convergence')


plt.figure(6)
sns.relplot(x="iter", y="reward", hue = "Algorithm", kind="line", 
            data=df2[df2['Algorithm'] == "Q-Learning Iteration"],legend = False,palette = "Set2")
plt.title('Easy Grid World Q-learning Reward')
plt.ylim(0, 100)

```
